### Web Scraping

Para obter os dados dos sites oficiais das Olímpiadas e Paralimpíadas, assim como da Wikipedia, vamos utilizar as bibliotecas `requests` e `Beautiful Soup`. Para processá-los e transformá-los em arquivos csv que facilitem nossa análise, vamos usar `pandas` e `io`.

--- 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import io
```

#### 1 - Acessando dados da Wikipedia

Primeiro vamos extrair o quadro de medalhas das Olimpíadas 

```python
wiki_url = "https://pt.wikipedia.org/wiki/Quadro_de_medalhas_dos_Jogos_Olímpicos"
h2_id = "Quadro_de_medalhas_dos_Jogos_Olímpicos_de_Verão"

headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (elia.dutra.sena@academico.ufpb.br)"}

response = requests.get(wiki_url, headers= headers).text
soup = BeautifulSoup(response, "html.parser")

span_titulo = soup.find(id= h2_id)
tabela_html = span_titulo.parent.find_next_sibling("table") 

html_string = str(tabela_html)
```

```python
df_list_olympics = pd.read_html(io.StringIO(html_string))
df_olympics = df_list_olympics[0]

df_olympics
```

Esse comando converte a tabela extraída para CSV

```python
df_olympics.to_csv("medalhas_olimpiadas_wikipedia.csv", index= False, sep= ',', encoding= 'utf-8-sig')
```

Agora, vamos fazer o mesmo para as Paralimpíadas

```python
wiki_url = "https://pt.wikipedia.org/wiki/Quadro_de_medalhas_dos_Jogos_Paralímpicos"
table_id = "mwEg"

headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (elia.dutra.sena@academico.ufpb.br)"}

response = requests.get(wiki_url, headers= headers).text
soup = BeautifulSoup(response, "html.parser")

tabela_html = soup.find(id= table_id)

html_string = str(tabela_html)
html_corrigido = html_string.replace('"5;"', '"5"')
```

```python
df_list_paralympics = pd.read_html(io.StringIO(html_corrigido))
df_paralympics = df_list_paralympics[0]

df_paralympics
```

E por fim, também convertemos a tabela para CSV

```python
df_paralympics.to_csv("medalhas_paralimpiadas_wikipedia.csv", index= False, sep= ',', encoding= 'utf-8-sig')
```
