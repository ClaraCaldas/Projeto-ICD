**Dados Econômicos (PIB 2024)**

Para a análise de correlação econômica, utilizamos os dados de PIB Nominal referentes ao ano de 2024.

Nota sobre a fonte: Os dados foram extraídos do portal Worldometer, que atua como um agregador de estatísticas globais. Embora a coleta não tenha sido realizada diretamente no portal do Banco Mundial, a base de dados do Worldometer para este recorte utiliza as métricas e dados oficiais do Banco Mundial do ano de 2024. Essa fonte foi escolhida pela estrutura da tabela, que favorece a realização do web scraping.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import io

url = "https://www.worldometers.info/pt/pib/pib-por-pais/?source=wb&region=worldwide&year=2024&metric=nominal"
headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (luis.oliveira@academico.ufpb.edu.br)"}

response = requests.get(url, headers=headers)
response.encoding = 'utf-8' 

soup = BeautifulSoup(response.text, "html.parser")
tabela_html = soup.find("table")

df_pib = pd.read_html(io.StringIO(str(tabela_html)))[0]
df_pib['País'] = df_pib['País'].str.strip()
df_pib
```

Gera o cvs

```python
df_pib.to_csv("pib_mundial_2024_worldometers.csv", index=False, sep=',', encoding='utf-8-sig')
```
