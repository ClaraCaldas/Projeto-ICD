Importa as bibliotecas para fazer o scraping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from io import StringIO
```

Acessa o link onde os dados serão coletados e cria a tabela

```python

wiki_url = "https://pt.wikipedia.org/wiki/Lista_de_países_por_Índice_de_Desenvolvimento_Humano"
table_id = "mw3Q"

headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (mariaeloisasilva498@gmail.com)"}

response = requests.get(wiki_url, headers= headers).text
soup = BeautifulSoup(response, "html.parser")

tabela_idh = soup.find('table', attrs={'id': table_id})
df = pd.read_html(StringIO(str(tabela_idh)))[0]

print(df.head())
```

Arruma as colunas para deixar a leitura mais clara e cria o csv

```python
df_list = pd.read_html(StringIO(str(tabela_idh)))
df = df_list[0]

df.rename(columns={'Δ': 'Mudança desde 2015'}, inplace=True)

df.to_csv("dadoss_idh_wikipedia.csv", index= False, sep= ',', encoding= 'utf-8-sig')
```
