**Jogos Paralímpicos de Verão de 1960 - Roma**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import io
```

```python
url_1960 = "https://www.paralympic.org/rome-1960/results/medalstandings"

headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (elia.dutra.sena@academico.ufpb.br)"}

response = requests.get(url_1960, headers= headers).text
soup = BeautifulSoup(response, "html.parser")

tabela_html = soup.find("table")

df_1960 = pd.read_html(io.StringIO(str(tabela_html)))[0]

df_1960 = df_1960.iloc[:, :7]
df_1960.columns = ['Ordem', 'País', 'CPN', 'Ouro', 'Prata', 'Bronze', 'Total']

df_1960 = df_1960.fillna(0)
df_1960 = df_1960[df_1960['País'].str.contains('Total|TOTAL') == False]

colunas_num = ['Ordem', 'Ouro', 'Prata', 'Bronze', 'Total']
df_1960[colunas_num] = df_1960[colunas_num].astype(int)

df_1960['País'] = df_1960['País'].str.strip()

df_1960
```

```python
df_1960.to_csv("medalhas_roma_1960_Para.csv", index=False, sep=',', encoding='utf-8-sig')
```
