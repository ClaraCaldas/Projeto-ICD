```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from io import StringIO
```

```python
wiki_url = "https://pt.wikipedia.org/wiki/Quadro_de_medalhas_dos_Jogos_Olímpicos"
h2_id = "mwDSg"

headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (mariaeloisasilva498@gmail.com)"}

response = requests.get(wiki_url, headers= headers).text
soup = BeautifulSoup(response, "html.parser")

sem_medalhas = soup.find('table', attrs={'id': h2_id})
df = pd.read_html(StringIO(str(sem_medalhas)))[0]
```

```python
df_list = pd.read_html(StringIO(str(sem_medalhas)))
df = df_list[0]


df_limpo = df.iloc[:, :2] 

print(df_limpo.columns.tolist())
display(df_limpo.head())
df = df_limpo

df.rename(columns={'Unnamed: 1': 'Número de participações'}, inplace=True)
df
```

```python
df.to_csv("paises_olimpicos_sem_medalhas_wikipedia.csv", index= False, sep= ',', encoding= 'utf-8-sig')
```

Paises paralimpicos

```python
wiki_url = "https://pt.wikipedia.org/wiki/Quadro_de_medalhas_dos_Jogos_Paralímpicos"
table_id = "mwCgk"

headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (mariaeloisasilva498@gmail.com)"}

response = requests.get(wiki_url, headers= headers).text
soup = BeautifulSoup(response, "html.parser")

tabela_html = soup.find(id= table_id)

html_string = str(tabela_html)
html_corrigido = html_string.replace('"5;"', '"5"')
```

```python
df_list_no_paralympics = pd.read_html(StringIO(html_corrigido))
df_no_paralympics = df_list_no_paralympics[0]


df_no_paralympics.columns = df_no_paralympics.columns.get_level_values(-1)

df_limpo = df_no_paralympics.iloc[:, :2] 

print(df_limpo.columns.tolist())
display(df_limpo.head())

df_limpo.columns = ['Time (Código IPC)', 'Número de participações']

df_no_paralympics = df_limpo

df_no_paralympics
```

```python
df.to_csv("paises_paralympicos_sem_medalhas_wikipedia.csv", index= False, sep= ',', encoding= 'utf-8-sig')
```
