**Jogos Olímpicos de Verão de 1960 - Roma**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import io

url_1960 = "https://pt.wikipedia.org/wiki/Quadro_de_medalhas_dos_Jogos_Olímpicos_de_Verão_de_1960"
headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (luis.oliveira@academico.ufpb.edu.br)"}

response = requests.get(url_1960, headers=headers)
response.encoding = 'utf-8'

soup = BeautifulSoup(response.text, 'html.parser')
tabela_html = soup.find("table", {"class": "wikitable"})

df_1960 = pd.read_html(io.StringIO(str(tabela_html)))[0]

df_1960 = df_1960.iloc[:, :6]
df_1960.columns = ['Ordem', 'País', 'Ouro', 'Prata', 'Bronze', 'Total']

df_1960 = df_1960.fillna(0)
df_1960 = df_1960[df_1960['País'].str.contains('Total|TOTAL') == False]

colunas_num = ['Ouro', 'Prata', 'Bronze', 'Total']
df_1960[colunas_num] = df_1960[colunas_num].astype(int)

df_1960['País'] = df_1960['País'].str.strip()

df_1960
```

```python
df_1960.to_csv("medalhas_roma_1960_oficial.csv", index=False, sep=',', encoding='utf-8-sig')
```
