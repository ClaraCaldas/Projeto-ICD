# Análise de Dados: O "Efeito Casa" nos Jogos Olímpicos (1960 - 2024)

Um dos nossos questionamentos diz respeito ao impacto de sediar os Jogos Olímpicos. O país sede realmente ganha mais medalhas quando joga em casa? Para responder a essa pergunta, coletamos dados históricos do quadro de medalhas desde a edição de Roma (1960) até Paris (2024), analisando o desempenho dos países em três momentos: **4 anos antes de sediar, no ano em que sediaram, e 4 anos depois**.

---

## 1.Evolução Histórica

Nossa primeira abordagem foi analisar o volume bruto de medalhas. No gráfico abaixo, agrupamos o desempenho de cada país sede. A barra azul escura representa a edição em que o país sediou os jogos.

![Gráfico de Evolução das Medalhas](grafico_efeito_casa_completo_blues.png)

**Destaques da Análise Visual:**
* **A expressiva evolução da Espanha:** O exemplo mais claro do Efeito Casa ocorreu em Barcelona (1992). A Espanha saltou de 4 medalhas na edição anterior para  22 medalhas jogando em casa.
* **O Domínio Chinês:** Em Pequim (2008), a China atingiu impressionante marca de 100 medalhas jogando em casa, consolidando-se definitivamente como uma potência olímpica.
* **As Exceções de 1980 e 1984:** Os "buracos"   nestes anos são explicados por fatores geopolíticos. O boicote liderado pelos EUA aos jogos de Moscou (1980) e o boicote soviético aos jogos de Los Angeles (1984) distorceram os números de "edição anterior" e "posterior" para essas nações.

---

## 2. A Prova Estatística: Correlação de Desempenho

 aplicamos o cálculo de **Correlação** comparando a edição anterior com a edição sede (excluindo os anos de boicote para não gerar divisão por zero).

 ### Por que excluímos os Estados Unidos (1984) da correlação?

Se observar o gráfico de dispersão com atenção, vai notar que deixamos a edição de 1984 de fora dessa etapa do cálculo. O motivo para isso é tanto histórico quanto matemático.

**O contexto do boicote:**
Em 1980, os Estados Unidos boicotaram as Olimpíadas de Moscou. Por causa disso, o número de medalhas deles na edição "Anterior" foi igual a zero. Quatro anos depois, sediando os jogos em Los Angeles (1984), eles pularam de 0 para 174 medalhas.

**O impacto nos dados:**
Na análise dos dados, um salto tão extremo (de 0 para 174) é o que chamamos de *outlier*, ou seja, um ponto totalmente fora da curva. Esse resultado aconteceu por uma questão política excepcional, e não por uma evolução esportiva. 

Se a gente deixasse esse dado na conta, ele iria distorcer toda a matemática do gráfico, puxando a linha de tendência para baixo e atrapalhando a análise do desempenho dos outros países.

**Conclusão:**
 Tirando esse evento isolado do cálculo, nós garantimos que a nossa correlação de 0,94 seja confiável e reflita o verdadeiro impacto esportivo de ser o país sede.

![Gráfico de Correlação](grafico_correlacao_linha.png)

**Os Resultados:**
* **Taxa de Sucesso de 93,3%:** Em 14 das 15 edições analisadas, o país sede **aumentou** seu número de medalhas. 
* **Correlação de 0,94:** A proximidade das bolinhas com a linha tracejada vermelha prova que o aumento de desempenho é previsível e recorrente.
* **Aumento Médio de 134%:** Em média, os países mais que dobram o seu total de medalhas quando são sede.
* **A Exceção:** O único país a não superar a própria marca foram os EUA em Atlanta (1996), com uma leve queda de 108 para 101 medalhas. Contudo, eles ainda lideraram o quadro geral daquele ano.

---

## 3. Conclusão: Por que o Efeito Casa é Real?

Os dados confirmam que o "Efeito Casa" é um fenômeno estatisticamente comprovado. Isso ocorre por uma combinação de fatores:

1. **Vagas Garantidas:** O país sede classifica atletas automaticamente para quase todas as modalidades, sem passar por eliminatórias globais, aumentando as chances matemáticas de pódio.
2. **Fator Físico e Psicológico:** Atletas locais não sofrem com fuso horário ou clima, contam com o apoio da torcida e conhecem a fundo as instalações esportivas.
3. **O Legado do Investimento:** Sediar os jogos exige um ciclo prévio de 8 anos de injeção massiva de verbas no esporte. Como reflexo, vemos nos gráficos que países como **Grã-Bretanha (após 2012)** e **Brasil (após 2016)** conseguiram manter um patamar elevado na edição seguinte, provando que a infraestrutura deixa um legado duradouro para as gerações futuras.

## 4. O Efeito Casa se repete nas Paralimpíadas?

Depois de comprovar o impacto de ser sede nos Jogos Olímpicos, a nossa próxima etapa foi aplicar exatamente a mesma metodologia para os Jogos Paralímpicos. O objetivo aqui é entender se o investimento em esportes adaptados pelo país anfitrião gera o mesmo tipo de salto no quadro de medalhas.

### Evolução Histórica Paralímpica

Separamos o desempenho das sedes nas Paralimpíadas usando a mesma lógica: comparando os resultados 4 anos antes, no ano do evento, e 4 anos depois.

![Gráfico de Evolução - Paralimpíadas](grafico_efeito_casa_paralimpiadas.png)

Como podemos ver no gráfico, o padrão se repete de forma até mais agressiva em alguns casos. Historicamente, os esportes paralímpicos costumam sofrer com falta de visibilidade e patrocínio. Quando um país é escolhido como sede, há uma injeção obrigatória de recursos em infraestrutura e acessibilidade que muitas vezes transforma completamente o desempenho daquela nação. 

### A Correlação Matemática nos Jogos Paralímpicos

Para manter a consistência do nosso estudo, rodamos o gráfico de dispersão com a linha de tendência para esses dados. Novamente, removemos anomalias e anos sem registros prévios para não comprometer a matemática.

![Gráfico de Correlação - Paralimpíadas](grafico_correlacao_linha_paralimpiadas.png)

O resultado do cálculo de correlação reforça a nossa tese. A linha de regressão mostra uma progressão clara: quem já tem um nível esportivo alto, dispara ao sediar os jogos; quem tem um nível menor, também ganha um impulso visível. 

Isso confirma que o "Efeito Casa" não é uma exclusividade das Olimpíadas convencionais. Ele é uma regra do esporte de alto rendimento como um todo, impulsionado pelo investimento, pela torcida e pelo fator de competir em casa.

### Uma curiosidade nos dados: Por que o volume de medalhas é tão maior?

Se você reparou bem nos gráficos, os números absolutos das Paralimpíadas chamam muita atenção. Em 1984, os Estados Unidos conquistaram **397 medalhas**, e países como a China costumam ultrapassar a marca de 200 pódios com facilidade. Nas Olimpíadas convencionais, passar de 100 medalhas já é um feito histórico. 

**Por que essa diferença tão grande acontece?**

A resposta está no número de provas disputadas, que é infinitamente maior nas Paralimpíadas devido ao sistema de **Classificação Funcional**. 

Para garantir que a competição seja justa, os atletas são divididos em "classes" de acordo com o tipo e o grau de sua deficiência (visual, física, intelectual, etc.). 

**Um exemplo prático:**
Nas Olimpíadas tradicionais, a prova dos 100 metros rasos no atletismo distribui apenas dois jogos de medalhas (um pódio masculino e um feminino). Já nas Paralimpíadas, existem **dezenas de provas diferentes de 100 metros rasos**. Há uma prova exclusiva para atletas cegos (com guias), outra para competidores em cadeiras de rodas, outra para amputados, e assim por diante.

Isso multiplica exponencialmente a quantidade total de medalhas em disputa ao longo do evento. Consequentemente, quando o "Efeito Casa" acontece e o país sede investe pesado no esporte adaptado, o salto nos gráficos paralímpicos parece muito mais agressivo do que nos gráficos olímpicos.
