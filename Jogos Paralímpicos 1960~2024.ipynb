**Jogos Paralímpicos de Verão de 1960 - Italia**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import io

url_para_1960 = "https://pt.wikipedia.org/wiki/Jogos_Paral%C3%ADmpicos_de_Ver%C3%A3o_de_1960"
headers = {"User-Agent": "ProjetoAnaliseDadosOlimpicos/1.0 (luis.oliveira@academico.ufpb.edu.br)"}

response = requests.get(url_para_1960, headers=headers)
response.encoding = 'utf-8'

soup = BeautifulSoup(response.text, 'html.parser')
tabela_html = soup.find("table", {"class": "wikitable"})

df_para_1960 = pd.read_html(io.StringIO(str(tabela_html)))[0]

df_para_1960 = df_para_1960.iloc[:, :6]
df_para_1960.columns = ['Ordem', 'País', 'Ouro', 'Prata', 'Bronze', 'Total']

df_para_1960 = df_para_1960.fillna(0)
df_para_1960 = df_para_1960[df_para_1960['País'].str.contains('Total|TOTAL') == False]

colunas_num = ['Ouro', 'Prata', 'Bronze', 'Total']
df_para_1960[colunas_num] = df_para_1960[colunas_num].astype(int)

df_para_1960['País'] = df_para_1960['País'].str.strip()

df_para_1960.to_csv("paralimpicos_roma_1960.csv", index=False, sep=',', encoding='utf-8-sig')

df_para_1960
```
