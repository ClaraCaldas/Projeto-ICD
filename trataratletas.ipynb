```python
import pandas as pd
```

Pega o arquivo com os dados dos atletas e lê

```python
df = pd.read_csv('athlete_events.csv', encoding='ISO-8859-1')
display(df.head())
```

Organiza o arquivo tirando as linhas que têm atletas que competiram nas olimpíadas de inverno

```python
df_organizado = df.drop(df[df['Year'] < 1960].index)
df_organizado = df[df["Season"] != "Winter"]


df_organizado.to_csv('athlete_events_arrumado.csv', index=False, encoding='ISO-8859-1')
df_organizado.head()
```
