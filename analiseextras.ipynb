**Eficiência da Delegação (Medalhas por Atleta Enviado)**


​A Pergunta: Qual país é mais eficiente? Quem envia menos atletas para ganhar uma medalha?


​A Comparação: O Brasil enviou 302 atletas para Tóquio e ganhou 21 medalhas. Compare essa taxa de conversão com países como EUA (muitos atletas) ou Cuba (menos atletas, mas alta eficiência histórica no boxe).
​Dado Extra: Tamanho da delegação (Team Size).


```python
import pandas as pd
```

```python
df_atletas = pd.read_csv('athlete_events_arrumado.csv', encoding='ISO-8859-1')
display(df_atletas.head())
```

```python
df_atletas.info()
display(df_atletas.describe())
display(df_atletas.shape) # número de linhas e colunas
display(list(df_atletas.columns))  # lista de colunas
```

Para cada ano de 1960 a 2020, ele filtra o dataframe para obter apenas os dados daquele ano, depois agrupa por NOC (Comitê Olímpico Nacional) e Year, contando o número de atletas para cada combinação. O resultado é armazenado em um dicionário onde a chave é o ano e o valor é um dataframe com as contagens de atletas por NOC para aquele ano.

```python
df_delegacao = {}

for i in range(1960, 2021, 4):
    df_ano = df_atletas[df_atletas['Year'] == i]
    df_delegacao[i] = df_ano.groupby(['NOC', 'Year']).size().reset_index(name='Tamanho Delegação')
display(df_delegacao[1960])

df_delegacaototal = pd.concat(df_delegacao.values(), ignore_index=True)
display(df_delegacaototal.head())
```
